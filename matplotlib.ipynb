```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.2)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
plt.savefig('pyplot1.png', dpi=450)
```

```python
from pylab import*
x= arange(0,10,0.2)
y = sin(x)
plot(x,y)
savefig('pylab1.png', dpi=450)
```

```python
fig = plt.figure()
fig.set_size_inches((8,5,4))
```

```python
#single axes - position values are 0-1
left, bottom, width, height = .1, .2, .7, .5
ax = fig.add_axes((left, bottom, width, height))
ax.plot(x,y)
ax2 = fig.add_axes((.9, .9, .1, .1))
ax2.plot(x,y)
fig
```

```python
# 3 axes- 1 row 3 cols(grid)
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax in axes:
    ax.plot(x,y)
```

```python
#1 axes - 1 row 2 cols 2nd postion
ax = plt.subplot(122)
ax.plot(x, y)
```

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='r', linewidth=3, linestyle='--')
fig.savefig('img/pyplot2.png', dpi=3000)
```
